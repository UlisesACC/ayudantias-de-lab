```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

data= pd.read_csv('AAPL.csv')
# Tomar la columna de los días que han pasado
X = data.iloc[:, 0].values
# Tomar el cierre de la bolsa
Y = data.iloc[:, 3].values

# Test será el 20% de los datos
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Entrenando el modelo
model = LinearRegression()
model.fit(X_train.reshape(-1, 1), Y_train)

# Predicción del modelo
Y_pred = model.predict(X_test.reshape(-1, 1))

# Calculamos el accuracy
accuracy = r2_score(Y_test, Y_pred)
print('Accuracy:', accuracy)

# Graficar los datos
plt.figure(figsize=(12, 6))
plt.scatter(X_test, Y_test, color='blue', label='Real')
plt.plot(X_test, Y_pred, color='orange', label='Predicción')
plt.title('Predicción de precios de acciones')
plt.legend()
plt.show()
```
